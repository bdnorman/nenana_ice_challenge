```python
import pandas as pd
```

```python
df = pd.read_csv("Shortwave Radiation Sum.csv", skiprows=2)
```

```python
df = df.rename(columns={"shortwave_radiation_sum (MJ/mÂ²)": "solar_radiation"})
df["time"] = pd.to_datetime(df["time"])
year = df[df.time.dt.year == 1989]
```

```python
year.set_index("time").solar_radiation.plot()
```

```python
monthly_avg = df.groupby(pd.Grouper(key="time", freq="M")).mean()

# Extract year and month from the index
monthly_avg["year"] = monthly_avg.index.year
monthly_avg["month"] = monthly_avg.index.month

# Filter for February (2) and March (3)
filtered_df = monthly_avg[monthly_avg["month"].isin([2, 3])]

# Pivot to create separate columns for Feb and March
result = filtered_df.pivot(index="year", columns="month", values="solar_radiation")

# Rename columns for clarityr
result.columns = ["feb_solar_radiation", "march_solar_radiation"]
```

```python
result.plot()
```
